### Finding the Index of an Item
we use the `.index() function` to get the index of an item


imagine we have a very long list, if you want to find **the index of the first occurance** of an item

for example:
```python
myList = [3, 8, 32, 3, 5, 6, 7, 7, 73, 2, 12, 'hello', 'hi', True, False, 21]
```

Now to find where 'hello' is we use
```python
myList.index('hello')
```

````python
myList = [3, 8, 32, 3, 5, 6, 7, 7, 73, 2, 12, 'hello', 'hi', True, False, 21]
myList.index('hello')
````

Remeber `.index()` is going to return the index of the ***`FIRST OCCURENCE of the item if the item has duplicate`***

In the above example, we know that we have the item 3 repeated 2 times(at index 0 and 3), so when we use `.index() `it will give us the index of first occurence of 3(which is 0)

````python
myList.index(3) #0, 3
````

````python
myList.index(7) # 6, 7
````

### Slicing
* Slicing is *`breaking list into subpart by using indices`*
* To do slicing we use `[starting_index: end_index]` , remember **`the value at end_index is not included in the slice`**

`Example`

```python
l = [10, 43, 2, 7]
l[0:2]
```

here we are slicing the values from index 0 to index 1 so we get `[10, 43]`

````python
l = [10, 43, 2, 7]
l[0:2] #0, 1 stop
````

````python
x = [1, 'hello', 0, 'world']
x[1:3] #1, 2,  ['hello', 0]
````

````python
evens = [i for i in range(2, 100, 2)]
len(evens)
````

````python
evens
````

````python
evens[4:8] #4, 5, 6, 7
````

````python
evens[4:] #from idnex 4 to the end
````

````python
evens[:8] #from start to index 7
````

if want everything ***from certain index up until the end***, we just use `[start_index:]`, we don't need to write the end_index

````python
x = [1, 'hello', 0, 'world']
x[2:] # this means we want everything starting from index 2
````

if want ***from the beginning until certain index***, we use `[:end_index]`, we don't need to write the beginning

````python
x[:2] #this means we want everything from the beginning till index 1
````
