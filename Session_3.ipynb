

````python
#Example of if statment
if True:
    print('Hello World!')
````

````python
if False:
    print('Hello, World!')
````

````python
a = 33
b = 200
if b > a: #True
  print("b is greater than a")
````

````python
# with conditional
if 1 == 2: #False
    print('Hello World!')
````

````python
# with else condition
if 1 == 1: #False
    print('Hello World!')
else:
    print('Goodbye world!')
````

````python
# Else is a catch-all
if 1 == 2:
    print('Hello World!')
else:
    print('Goodbye world!')   #Else is executed when the 'if' condition is false
````

````python
# elif only triggers on the condition
x = 1
if x == 0: #false
    print('x=0')
elif x == 2: #false
    print('x=2')
else:
    print("x is neither 0 nor 2")
````

````python
'''
    1. Take three int values from user and print greatest among them.
    x, y, z
    x > y... x > z
    y > x, y > z..
    else: z
'''
x = int(input("enter the first number? "))
y = int(input("enter the second number? "))
z = int(input("enter the third number? "))

if x > y and x > z:
  print(f"the largest number is {x}")
elif y > x and y > z:
  print(f"the largest number is {y}")
else:
  print(f"the largest number is {z}")


````
