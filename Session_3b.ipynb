## For Loops
* **`For loops iterate over a given sequence`**

syntax
```python
for varaible in sequence:
        do something
```

***Remember: sequence means list, string...***

```python
primes = [2, 3, 5, 7]
for variable in primes: #variable takes values in primes one by one in each iteration
  print(variable) #2 3 5 7

```

````python
#let's say we have list named primes
primes = [2, 3, 5, 7]

#we want to go item by item inside the list using for loop
for variable in primes: #variable will take the values of prime one by one
    print(variable)
````

In the above code
```python
for i in primes
```
mean the variable `'i' iterates over all items in our list primes one by one.`

````python
name = "Ethiopia" #this is our sequence
for i in name: #going one by one inside the string
    print(i)
````

### Range function
Range function. it can take 1, 2 or 3 arguments

when range has 1 argument(call it n), it is like looping over the list from 0 to n-1

syntax
```python
range(number)
# this gives us list of numbers from 0 till number -1
#for example
range(10) #mean [0, 1, 2, 3,...9]
```

````python
list(range(10))
````

````python
list(range(7))
````

````python
list(range(100))
````

#### Range with 2 arguments
syntax
```python
range(start, end)
#means from start to end-1
range(2, 6) #mean [2, 3, 4, 5]
```


````python
list(range(2,7))
````

````python
list(range(1,4))
````

#### Range with 3 arguments
syntax
```python
range(start, end, skip)
# means from start to end-1 skipping some amount
range(2,9,3) # gives [2, 5, 8]
```

````python
list(range(2,9,3)) #2, 5,8
````

````python
list(range(1,9,2)) # 1, 3, 5, 7
````

````python
# For loops can iterate over a sequence of numbers using the "range" function

# Prints out the numbers 0,1,2,3,4
for x in range(5):
    print(x)


# Prints out 3,4,5
for x in range(3, 6):
    print(x)

# Prints out 3,5,7
for x in range(3, 8, 2):
    print(x)
````
