## For Loops
* **`For loops iterate over a given sequence`**

syntax
```python
for varaible in sequence:
        do something
```

***Remember: sequence means list, string...***

````python
#let's say we have list named primes
primes = [2, 3, 5, 7]
#we want to go item by item inside the list using for loop
for variable in primes: #i will take the values of prime one by one
    print(variable)
````


### Range function
Range function. it can take 1, 2 or 3 arguments

when range has 1 argument(call it n), it is like looping over the list from 0 to n-1

syntax
```python

#for example

```

````python
list(range(10))
````

````python
list(range(7))
````

#### Range with 2 arguments
syntax


## While Loops
While loop does the same thing as for loop!
* **The main difference is, they are often used when we do not know how many times the iteration will happen.**

* While loops repeat as long as a certain `boolean condition is met`.

syntax
```python
while CONDITION:
  Do Something

#Example:
count = 0
while count < 5: #the loop continues as long as our variable count is less than 5
    print(count)
    count += 1  # This is the same as count = count + 1

```

````python
count = 0
while count < 5:
    print(count)
    count += 1
````

### Caution: when using while loop, you should be very careful about stopping condition!
`If the stopping condition is never met, the loop goes forever!!`

````python
'''
counter = 2
while counter > 0:
  print(i)
'''
#in the above loop counter is never updated and it's always 2, which means the condition counter > 0 is alwasy true! so the loop goes forever!
````

````python
# Print i as long as i is less than 6:
i = 1
while i < 6:
  print(i)
  i += 1 #increment(updating i so that it goes to the stopping stage)
````

````python
fruits = ['orange', 'banana', 'apple', 'pineapple']
#printing all fruits
y = 0
while y < len(fruits):
    print(fruits[y])
    y += 1
````

### Break and Continue
* break: as it's name suggests stop or breaks while or for loop
* continue: will skip the current block and return to the 'while' or 'for' loop statement

````python
# Prints out 0,1,2,3,4
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break
````

````python
# Prints out only odd numbers - 1,3,5,7,9
for x in range(10):
    # Check if x is even
    if x % 2 == 0:
        continue
    print(x)
````

````python
# What if I only want to print orange?
fruits = ['orange', 'banana', 'apple', 'pineapple']
i = 0
while i < len(fruits):
    print(fruits[i])
    if i > 0: #what is wrong?
        break
    i += 1 # what happens if we comment this out?
````

````python
# What if I only want to print orange?
fruits = ['orange', 'banana', 'apple', 'pineapple']
i = 0
while i < len(fruits):
    print(fruits[i])
    i += 1
    if i > 0:
        break
````

````python
# What if I only want to print pineapple?
fruits = ['orange', 'banana', 'apple', 'pineapple']
i = 0
while i < len(fruits):
    if i == len(fruits)  - 1:
        print(fruits[i])
    i += 1
````
